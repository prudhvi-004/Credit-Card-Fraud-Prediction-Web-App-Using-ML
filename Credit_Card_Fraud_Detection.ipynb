**Under Sampling**

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
df = pd.concat([legit_sample,fraud],axis=0)
```

**Splitting the Data**

```python
X = df.drop('Class',axis=1)
Y = df['Class']
```

```python
print(X)
```

```python
print(Y)
```

**Training and Testing**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

**Model Evaluation**

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
def evaluate(true,predicted):
  acc = accuracy_score(true,predicted)
  pre = precision_score(true,predicted)
  rec = recall_score(true,predicted)
  f1 = f1_score(true,predicted)

  return acc,pre,rec,f1
```

**Model Training**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier

from sklearn.model_selection import cross_val_score
```

```python
models = [LogisticRegression(),SVC(),KNeighborsClassifier(),DecisionTreeClassifier(),RandomForestClassifier(),XGBClassifier(),GradientBoostingClassifier(),AdaBoostClassifier()]
```

```python
for model in models:
  model.fit(X_train,Y_train)

  Y_train_pred = model.predict(X_train)
  Y_test_pred = model.predict(X_test)

  acc,pre,rec,f1 = evaluate(Y_train,Y_train_pred)

  print("Model Name:",model)

  print()

  print("Training Results:")
  print("Accuracy:",acc)
  print("Precision:",pre)
  print("Recall:",rec)
  print("F1 Score:",f1)

  print()

  print("Testing Results:")
  print("Accuracy:",acc)
  print("Precision:",pre)
  print("Recall:",rec)
  print("F1 Score:",f1)

  print()
  cv_score = cross_val_score(model,X,Y,cv=5)
  print("Cross Validation:",cv_score)
  print("Cross Validation Score:",cv_score.mean())

  print("---------------------------------------------------------------------------")
```

**Hyper Parameter Tuning**

```python
gb_params = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [3, 5, 7],
    "subsample": [0.6, 0.8, 1.0],
    "max_features": ["sqrt", "log2",None]
}
```

```python
random_cv = [
    ("GB", GradientBoostingClassifier(), gb_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model_params = {}

for name,model,params in random_cv:
    random = RandomizedSearchCV(estimator=model,
                                param_distributions=params,
                                verbose = 2,
                                n_jobs= -1,
                               n_iter=10,
                               )
    random.fit(X_train,Y_train)
    model_params[name] = random.best_params_
for model_name in model_params:
    print(f"--------------------Best Paarams for {model_name}-------------------")
    print(model_params[model_name])
```

```python
model = random.best_estimator_
```

```python
print(model)
```

```python
model.fit(X_train,Y_train)

test_pred = model.predict(X_test)

acc = accuracy_score(Y_test,test_pred)

print(acc)
```
